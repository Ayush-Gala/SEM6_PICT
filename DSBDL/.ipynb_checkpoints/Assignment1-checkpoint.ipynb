```python
import pandas as pd
import numpy as np

# using read_csv function to read the dataset
df = pd.read_csv('melb_data.csv')
df.head()
```

```python
df.describe()
#talks about each value inside the dataset
#gives the values of quartiles and min/max of data
```

```python
print(df.shape)
```

```python
df.info()
#gives basic information of the dataset variables. Defines the data type, count and null status of the columns
```

```python
#getting the null values for each column in the dataset. 
#for loop iterates over the columns of the dataset, isna function checks for null values and sum function aggregates the null values.
for elem in df:
    print(elem, " - ", df[elem].isna().sum())
```

```python
df.dropna(axis=0, thresh=20)
#it is preferable to keep the threshold as the number of columns in the dataset.
# Otherwise the dataset may contain null values alog the axis.
#Keeping axis = 1 removes the entire column and is hence not desirable

```

```python
df.fillna(method='pad')
```
