```python
#detecting the outliers in the data set

df.plot.scatter(y="Reading Score", x="Math Score", s=4, figsize=(20,10))
```

```python
df.plot.scatter(y="Reading Score", x="Writing Score", s=4, figsize=(20,10))
```

```python
#removing the outliers from the dataset

df = df.drop(df[df['Math Score'] > 100].index, inplace=False)
df = df.drop(df[df['Writing Score'] > 100].index, inplace=False)
df = df.drop(df[df['Reading Score'] > 100].index, inplace=False)

df.plot.scatter(x="Reading Score", y="Writing Score", s=4, figsize=(20,10))
    
```
